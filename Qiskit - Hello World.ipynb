# Hello World in Qiskit

This examples simulates a simple bit and measures it. We begin by first loading in the library


```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit import Aer, transpile
import math
import qiskit
from qiskit.tools.visualization import plot_histogram
```

Got to check the correct version

```python
qiskit.__qiskit_version__
```

We construct a simple circuit of a single q-bit. We also suppy the initial value to state. The goal of cicuit is simple: Measure it

```python
 # Create a quantum circuit with one qubit
circ = QuantumCircuit(1) 
# Define initial_state as (1/sqrt(2))|0>+(1/sqrt(2))|1>
initial_state = [1/math.sqrt(2),1/math.sqrt(2)]   
circ.initialize(initial_state, 0)
# Place a measurement gate
circ.measure_all()
```

Qiskit provides a simple utility to draw the circuit.

```python
circ.draw('mpl')
```

Rearrange the circuit to meet the needs of the simulator/hardware (transpile) and simulate them. 

```python
simulator = Aer.get_backend('aer_simulator')
circ = transpile(circ, simulator)

# Run and get counts
result = simulator.run(circ).result()
counts = result.get_counts(circ)
plot_histogram(counts, title='Measured Values')
```
